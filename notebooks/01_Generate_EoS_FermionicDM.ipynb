# Fermionic Dark Matter

Equation of state in dimensionless units parameterized by the dimensionless Fermi momentum $z = k_F / m_f$:

$$\tilde{\varepsilon} = \frac{1}{8\pi^2} \left[ (2z^3 + z)\sqrt{1+z^2} - \text{arcsinh}(z) \right]$$

$$\tilde{p} = \frac{1}{24\pi^2} \left[ (2z^3 - 3z)\sqrt{1+z^2} + 3\text{arcsinh}(z) \right]$$

$$\tilde{n} = \frac{z^3}{3\pi^2}$$

Where, to recover the physical quantities scaled to the $4B$ units (as implemented in the numerical solver):

$$\varepsilon = \frac{m_f^4}{4B} \tilde{\varepsilon}, \quad P = \frac{m_f^4}{4B} \tilde{p}$$

$$n = \frac{m_f^3}{(4B)^{3/4}} \tilde{n}$$

```python
def generate_fdm_eos(m_f_MeV, B_val_MeV4, z_min=2e-4, z_max=100, points=1000):

    z = np.geomspace(z_min, z_max, int(points))
    
    term_sqrt = np.sqrt(1 + z**2)
    term_arcsinh = np.arcsinh(z)
    
    p_prime = (1 / (24 * np.pi**2)) * ((2*z**3 - 3*z) * term_sqrt + 3 * term_arcsinh)
    e_prime = (1 / (8 * np.pi**2))  * ((2*z**3 + z) * term_sqrt - term_arcsinh)
    n_prime = z**3 / (3 * np.pi**2)
    
    scale_pe = (m_f_MeV**4) / (4 * B_val_MeV4)
    scale_n = (m_f_MeV**3) / (4 * B_val_MeV4)**(0.75)
    
    df = pd.DataFrame({
        'z_dimensionless': z,
        'Energy Density': e_prime * scale_pe,
        'Pressure': p_prime * scale_pe,
        'Number_Density': n_prime * scale_n
    })
    
    return df
```
