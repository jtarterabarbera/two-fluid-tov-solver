# Bosonic Dark Matter

Equation of state in dimensionless units
$$
\tilde\epsilon = \tilde p ^{2/n} + \frac{n+2}{n-2}\tilde p
$$
Where, to recover physical units $\varepsilon = m_b^4 \ \tilde \varepsilon$, $P = m_b^4 \ \tilde P$

$$\tilde n = \tilde{p}^{2/n} \sqrt{ \frac{2n}{n-2} \tilde{p}^{1-2/n} + 1 }$$
where, to recover physical units $n=\ m_b^3 \tilde n$


```python
# ---------------------------------------------------------
# 1. Thermodynamic Functions
# ---------------------------------------------------------

def energy_density(pressure, n):
    return pressure ** (2/n) + ((n + 2) / (n - 2)) * pressure

def number_density(pressure, n):
    inner_bracket = ((2*n) / (n - 2) * pressure ** (1 - 2/n) + 1)
    return pressure ** (2/n) * np.sqrt(inner_bracket)

```

```python
# ---------------------------------------------------------
# 2. Get the EoS data csv file
# ---------------------------------------------------------

def get_csv_grid_EoS_BDM(n, m_b, B):

    B_val = B ** 4
    p_dimless = np.logspace(-20, 10, int(1e5)) 
    e_dimless = energy_density(p_dimless, n)
    n_dimless = number_density(p_dimless, n)

    
    grid_p_DM = p_dimless * m_b**4 / (4*B_val)
    grid_e_DM = e_dimless * m_b**4 / (4*B_val)
    grid_n_DM = n_dimless * m_b**3 / ((4 * B_val) ** (3/4))

    grid_eos = pd.DataFrame({
        'Energy Density': grid_e_DM,
        'Pressure': grid_p_DM,
        'Number Density': grid_n_DM
    })
    
    name_B = int(B)
    name_csv = f'../data/eos_library/bosonicDM_n_{n}_mb_{m_b}_B_{name_B}.csv'
    grid_eos.to_csv(name_csv, index=False)
    
    return None
```

```python
get_csv_grid_EoS_BDM(n=40, m_b=300, B=148)
get_csv_grid_EoS_BDM(n=4, m_b=300, B=148)
get_csv_grid_EoS_BDM(n=4, m_b=1000, B=148)
get_csv_grid_EoS_BDM(n=4, m_b=200, B=148)
get_csv_grid_EoS_BDM(n=40, m_b=200, B=148)
get_csv_grid_EoS_BDM(n=40, m_b=1000, B=148)
```
